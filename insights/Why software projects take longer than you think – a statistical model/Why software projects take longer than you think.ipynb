# Why software projects take longer than you think

### Abstract

It is a well known fact that programming tasks are hard to estimate. 

## Introduction (maybe not needed)

```python
# SETUP
import scipy.stats as stats
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter
import scipy.stats as st
%config InlineBackend.figure_format = 'retina'
from IPython.display import HTML, Image, Video
```

## Dataset

Dataset: [The SiP dataset](https://github.com/Derek-Jones/SiP_dataset). A dataset of effort estimates on different software projects at the Software in Partnership (SiP) in UK, a company that makes back office insurance systems. The dataset contains over 10 000 task estimates made by 22 developers on 20 projects between 2004â€“2014.


```python
ds = pd.read_csv("data.csv")

print("Rows:", len(ds))

ds.head()
```

```python
fig, ax = plt.subplots()
ax.scatter(ds.HoursEstimate, ds.HoursActual)

ax.set_title('Estimated vs. actual development time')
ax.set_xlabel('Estimated time (hours)')
ax.set_ylabel('Actual time (hours)')
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlim(0.1, 1000)
ax.set_ylim(0.1, 1000)
ax.xaxis.set_major_formatter(ScalarFormatter())
ax.yaxis.set_major_formatter(ScalarFormatter())
```

```python
from mpl_toolkits.mplot3d import Axes3D
xx, yy = np.mgrid[-1:3:100j, -1:3:100j]
positions = np.vstack([xx.ravel(), yy.ravel()])
values = np.vstack([np.log(ds.HoursEstimate), np.log(ds.HoursActual)])
kernel = st.gaussian_kde(values)
f = np.reshape(kernel(positions).T, xx.shape)
fig = plt.figure(figsize=(13, 7))
ax = fig.add_subplot(221, projection='3d')
surf = ax.plot_surface(xx, yy, f, rstride=1, cstride=1, cmap='coolwarm', edgecolor='none')
ax.set_xlabel('Estimated time (log hours)')
ax.set_ylabel('Actual time (log hours)')
ax.set_zlabel('PDF')
ax.set_title('Joint probabilty density')
ax.set_xlim(3, -1)
ax.set_ylim(3, -1)
fig.colorbar(surf, shrink=0.5, aspect=5) # add color bar indicating the PDF
ax.view_init(90, 90)

ax = fig.add_subplot(222, projection='3d')
surf = ax.plot_surface(xx, yy, f, rstride=1, cstride=1, cmap='coolwarm', edgecolor='none')
ax.set_xlabel('Estimated time (log hours)')
ax.set_ylabel('Actual time (log hours)')
ax.set_zlabel('PDF')
ax.set_title('Joint probabilty density')
ax.set_xlim(3, -1)
ax.set_ylim(3, -1)
fig.colorbar(surf, shrink=0.5, aspect=5) # add color bar indicating the PDF
ax.view_init(45, 90)
```

```python
actual_estimated_ratio = ds.HoursActual / ds.HoursEstimate
actual_estimated_ratio.hist(bins=100)

median = np.median(actual_estimated_ratio)
mean = np.mean(actual_estimated_ratio)
std = np.std(actual_estimated_ratio)
skew = stats.skew(actual_estimated_ratio)
kurtosis = stats.kurtosis(actual_estimated_ratio)
print(mean, std, skew, kurtosis)
```

```python
HTML("The average programming task takes <strong>%.02f</strong> many times more than estimated" % mean)
```

```python
HTML("With a standard deviation of <strong>%.02f</strong>" % std)
```

```python
HTML("... while the median programming task takes <strong>%.02f</strong> many times more than estimated" % median)
```

```python
estimated_actual_ratio = np.log(ds.HoursEstimate / ds.HoursActual)
estimated_actual_ratio.hist(bins=50)

mean = np.mean(estimated_actual_ratio)
std = np.std(estimated_actual_ratio)
skew = stats.skew(estimated_actual_ratio)
kurtosis = stats.kurtosis(estimated_actual_ratio)
print(mean, std, skew, kurtosis)
```
